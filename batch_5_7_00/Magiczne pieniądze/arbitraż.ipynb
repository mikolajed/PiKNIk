```python
PLNUSD = pd.read_csv("https://query1.finance.yahoo.com/v7/finance/download/PLNUSD=X?period1=1602884651&period2=1634420651&interval=1d&events=history&includeAdjustedClose=true")
USDEUR = pd.read_csv("https://query1.finance.yahoo.com/v7/finance/download/USDEUR=X?period1=1602884651&period2=1634420651&interval=1d&events=history&includeAdjustedClose=true")
EURPLN = pd.read_csv("https://query1.finance.yahoo.com/v7/finance/download/EURPLN=X?period1=1602884651&period2=1634420651&interval=1d&events=history&includeAdjustedClose=true")
```

```python
PLNUSD = PLNUSD.iloc[1:, 0:2]
USDEUR = USDEUR.iloc[1:, 0:2]
EURPLN = EURPLN.iloc[1:, 0:2]
```

```python
#PLNUSD, USDEUR, EURPLN
```

```python
labels = ['Date', 'PLNUSD', 'USDEUR', 'EURPLN']
PLNUSD = PLNUSD.to_numpy()
USDEUR = USDEUR.to_numpy()
EURPLN = EURPLN.to_numpy()
```

```python
data = [[date] for date in PLNUSD[:,0]]
data = np.concatenate((data, 
                       PLNUSD[:,1].reshape(-1, 1), 
                       USDEUR[:,1].reshape(-1, 1), 
                       EURPLN[:,1].reshape(-1, 1)), axis=1)
```

```python
#Handling missing values
data = [row for row in data if 
                            math.isnan(row[1])==False and
                            math.isnan(row[2])==False and 
                            math.isnan(row[3])==False]
```

```python
#To-do: build graph of assets conversions
class Graph:
    V = {}
    #For edges in E, multiply the edge value to move to another node
    E = {}
    
    def __init__(self, labels, relations):
        for i in range(len(labels)):
            self.V[i] = labels[i]
        for e in relations:
            if (e[0] in self.E):
                self.E[e[0]].append(e[1])
            else:
                self.E[e[0]] = [e[1]]
```

```python
edges = [[0, 1], [1, 2]]
G = Graph(labels[1:], edges)
```
